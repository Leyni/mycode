```python
from sklearn.datasets import load_iris
iris_data = load_iris()

print iris_data.DESCR
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

X_data = StandardScaler().fit_transform(iris_data.data)
Y_data = OneHotEncoder(sparse=False).fit_transform(iris_data.target.reshape(-1, 1))

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.3, random_state=0)

print X_train, Y_train
print X_test, Y_test
```

```python
import tensorflow as tf

x = tf.placeholder('float', [None, 4])
y = tf.placeholder('float', [None, 3])

w = tf.get_variable(
    name='w_weight', 
    shape=(4, 3), 
    initializer=tf.random_normal_initializer(stddev=0.1),
    regularizer=tf.contrib.layers.l2_regularizer(0.1)
)

b = tf.get_variable(
    name='b_weight', 
    shape=(1, 3), 
    initializer=tf.random_normal_initializer(stddev=0.1),
    regularizer=tf.contrib.layers.l2_regularizer(0.1)
)

logits = tf.matmul(x, w) + b
pred = tf.nn.softmax(logits)

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y, logits=pred))
optimizer = tf.train.AdamOptimizer(learning_rate=0.1)
train_op = optimizer.minimize(cost)
```

```python
from sklearn.utils import shuffle

init_global = tf.global_variables_initializer()
init_table = tf.tables_initializer()
init_local = tf.local_variables_initializer()

sess = tf.Session()
sess.run([init_global, init_table, init_local])

for i in range(80):
    loss = 0.0
    
    _, train_pred, train_loss = sess.run(
        [train_op, pred, cost],
        feed_dict={x: X_train, y: Y_train}
    )
    
    test_pred, test_loss = sess.run(
        [pred, cost],
        feed_dict={x: X_test, y: Y_test}
    )
    
    print("Epoch: %d, Train Logloss: %f, Test Logloss: %f" % (i, train_loss, test_loss))
```
